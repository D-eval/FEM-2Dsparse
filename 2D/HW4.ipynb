## 1

$$u = xy(1-x/2)(1-y)e^{x+y}$$

$$a(x,y) = 1$$

只须令

$$f = [(y-y^2)(1-x-x^2/2)+(x-x^2/2)(-3y-y^2)]e^{x+y}$$

$$ g = (x-x^2/2)(y-y^2)e^{x+y} $$

$$\kappa = 1$$

```python
import matplotlib.pyplot as plt
import numpy as np
from Dirichlet2D import solve_Dirichlet, plot_compare
# 定义边界
xy_boundary = np.array([[0,0], [1,0], [1,1], [0,1]])
# 网格半径
dh = 0.2

kappa = lambda xy: 1
g = lambda xy: (xy[0]-xy[0]**2/2)*(xy[1]-xy[1]**2)*np.exp(xy[0]+xy[1])
f = lambda xy: ((xy[1]-xy[1]**2)*(1-xy[0]-xy[0]**2/2)+(xy[0]-xy[0]**2/2)*(-3*xy[1]-xy[1]**2))*np.exp(xy[0]+xy[1])
u_real = g
Point2Value, all_point, all_tri = solve_Dirichlet(xy_boundary, kappa, g, f, dh, num_epoch=5000)
l1_loss, ax = plot_compare(Point2Value, u_real)
plt.savefig('./output/HW4_Q1.png')
plt.close()
```

![结果](./output/HW4_Q1.png)

```python
point2L1, ax = plot_compare(Point2Value, u_real)
sum_l1 = 0
for k,v in point2L1.items():
    sum_l1 += v.item()
print("l1 loss:", sum_l1)
```

## 2

$$
u = x^{2.5}\cos(y)
$$

令

$$
f = \cos y x^{0.5} (2.5*1.5 - x^2)
$$

$$
g = x^{2.5} \cos{y}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
from Dirichlet2D import solve_Dirichlet, plot_compare
# 定义边界
xy_boundary = np.array([[0,0], [1,0], [1,1], [0,1]])
# 网格半径
dh = 0.2

kappa = lambda xy: 1
g = lambda xy: xy[0]**(2.5)*np.cos(xy[1])
f = lambda xy: np.cos(xy[1])*xy[0]**(0.5)*(2.5*1.5-xy[0]**2)
u_real = g
Point2Value, all_point, all_tri = solve_Dirichlet(xy_boundary, kappa, g, f, dh, num_epoch=5000)
l1_loss, ax = plot_compare(Point2Value, u_real)
plt.savefig('./output/HW4_Q2.png')
plt.close()
```

![结果](./output/HW4_Q2.png)

```python
sum_l1 = 0
for k,v in l1_loss.items():
    sum_l1 += v.item()
print("l1 loss:", sum_l1)
```

## 3

令

$$
f = x^{0.5}e^{x+y}(\cos y 2.5(x+1.5)-(\sin{y} + \cos{y}))
$$

$$
g = x^{2.5} \cos{y}
$$

$$
\kappa = e^{x+y}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
from Dirichlet2D import solve_Dirichlet, plot_compare
# 定义边界
xy_boundary = np.array([[0,0], [1,0], [1,1], [0,1]])
# 网格半径
dh = 0.2

kappa = lambda xy: np.exp(xy[0]+xy[1])
g = lambda xy: xy[0]**(2.5)*np.cos(xy[1])
f = lambda xy: xy[0]**0.5 * np.exp(xy[0]+xy[1])*(np.cos(xy[1])*2.5*(xy[0]+1.5)-(np.sin(xy[1]) + np.cos(xy[1])))
u_real = g
Point2Value, all_point, all_tri = solve_Dirichlet(xy_boundary, kappa, g, f, dh, num_epoch=5000)
l1_loss, ax = plot_compare(Point2Value, u_real)
plt.savefig('./output/HW4_Q3.png')
plt.close()
sum_l1 = 0
for k,v in l1_loss.items():
    sum_l1 += v.item()
print("l1 loss:", sum_l1)
```
