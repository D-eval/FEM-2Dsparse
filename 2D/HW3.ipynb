我没有用到信息矩阵。我用了两个列表分别管理所有三角形和顶点，其中每个三角形可以获取顶点，也可以得到邻居三角形。因此可以通过是否拥有三个邻居三角形来判断是否是边界三角形。这样也能节约内存。

下面我来演示，对于任意的边界，我们逆时针在边界上取 $n$ 个点，得到一个 $(n,2)$ 的 boundary。然后选择一个 $\mathrm{d}h$，即可生成三角网格。我采用了德劳内剖分。

具体来说，先按照 $\mathrm{d} h$ 取 meshgrid 一些点，然后计算每个点环绕边界的角度，删去那些外部的点。对于内部的点，采用德劳内算法进行三角形剖分。

```python
import mesh_div
import numpy as np

boundary = np.array([[0,0], [1,0], [1,1], [0,1]]) # 对于正方形边界
# 注意，当选择较大边界时，要手动更改初始德劳内三角形
dh = 0.1
all_tri = mesh_div.get_tri(boundary, dh, need_disp=False)
```
