## 2


$$
\begin{aligned}
&
\sigma^{T} \nabla = f
\\&
u = 0 & \forall x \in \partial \Omega
\end{aligned}
$$

where

$$
\begin{aligned}
&
\epsilon = 
\begin{bmatrix}
\partial_1 u_1 & \frac{1}{2}(\partial_2 u_1 + \partial_1 u_2) \\
\frac{1}{2}(\partial_2 u_1 + \partial_1 u_2) & \partial_2 u_2
\end{bmatrix}
\\&
\sigma = (\lambda \nabla \cdot u) I + 2\mu \epsilon
\end{aligned}
$$

Then

$$
\sigma =
\lambda
\begin{bmatrix}
\partial_1 u_1 + \partial_2 u_2 &\\
& \partial_1 u_1 + \partial_2 u_2
\end{bmatrix}
+ \mu
\begin{bmatrix}
2\partial_1 u_1 & \partial_2 u_1 + \partial_1 u_2 \\
\partial_2 u_1 + \partial_1 u_2 & 2 \partial_2 u_2
\end{bmatrix}
$$

$$
\sigma^T \nabla = 
\lambda
\begin{bmatrix}
\partial_{11} u_1 + \partial_{12} u_2 \\
\partial_{22} u_2 + \partial_{12} u_1
\end{bmatrix}
+ \mu
\begin{bmatrix}
2\partial_{11} u_1 + \partial_{22} u_1 + \partial_{12} u_2 \\
2\partial_{22} u_2 + \partial_{11} u_2 + \partial_{12} u_1
\end{bmatrix}
=
\begin{bmatrix}
f_1 \\ f_2
\end{bmatrix}
$$

We consider this first equation

$$
\lambda (
    \partial_{11} u_1 +
    \partial_{12} u_2
) +
\mu (
    2 \partial_{11} u_1 +
    \partial_{22} u_1 +
    \partial_{12} u_2
) = f_1
$$

Weaky form

$$
\begin{aligned}
&
(\lambda + 2\mu)(<\partial_1 u_1, \partial_1 v>-<\partial_1 u_1, v>_{\partial_1}) +
\\&
(\lambda + \mu)(<\partial_2 u_2, \partial_1 v>-<\partial_2 u_2, v>_{\partial_1}) +
\\&
\mu (<\partial_2 u_1, \partial_2 v>-<\partial_2 u_1, v>_{\partial_2})
\\&
= -<f_1, v>
\end{aligned}
$$

Let

$v=\Phi_i$

$u_1 = \sum c_{1j} \Phi_j$

$u_2 = \sum c_{2j} \Phi_j$

Then

$$
\begin{aligned}
&
(\lambda + 2\mu) (
    \sum c_{1j} (<\partial_1 \Phi_j, \partial_1 \Phi_i> -
    <\partial_1 \Phi_j, \Phi_i>_{\partial_1}
    )
) +
\\&
(\lambda + \mu) (
    \sum c_{2j} (
        <\partial_2 \Phi_j, \partial_1 \Phi_i> -
        <\partial_2 \Phi_j, \Phi_i>_{\partial_1}
    )
) +
\\&
\mu (
    \sum c_{1j} (
        <\partial_2 \Phi_j, \partial_2 \Phi_i> -
        <\partial_2 \Phi_j, \Phi_i>_{\partial_2}
    )
)
\\&
= - <f_1, \Phi_i>
\end{aligned}
$$

and his brother


$$
\begin{aligned}
&
(\lambda + 2\mu) (
    \sum c_{2j} (<\partial_2 \Phi_j, \partial_2 \Phi_i> -
    <\partial_2 \Phi_j, \Phi_i>_{\partial_2}
    )
) +
\\&
(\lambda + \mu) (
    \sum c_{1j} (
        <\partial_1 \Phi_j, \partial_2\Phi_i> -
        <\partial_1 \Phi_j, \Phi_i>_{\partial_2}
    )
) +
\\&
\mu (
    \sum c_{2j} (
        <\partial_1 \Phi_j, \partial_1 \Phi_i> -
        <\partial_1 \Phi_j, \Phi_i>_{\partial_1}
    )
)
\\&
= - <f_2, \Phi_i>
\end{aligned}
$$


## 3

$$
\begin{bmatrix}
\partial_{1} u & \partial_{2} u
\end{bmatrix}

\begin{bmatrix}
c_{11} & c_{21} \\ c_{12} & c_{22}
\end{bmatrix}

\begin{bmatrix}
\partial_{1}  \\ \partial_{2} 
\end{bmatrix}

= f
$$

$$
\partial_1 (c_{11} \partial_1 u) +
\partial_2 (c_{21} \partial_1 u) +
\partial_1 (c_{12} \partial_2 u) +
\partial_2 (c_{22} \partial_2 u) = f
$$

$$
<c_{11}\partial_1 u, \partial_1 v> +
<c_{21}\partial_1 u, \partial_2 v> +
<c_{12}\partial_2 u, \partial_1 v> +
<c_{22}\partial_2 u, \partial_2 v> =
- <f, v>
$$

在题目中， 真实解$u=e^{x+y}$, $c=\begin{bmatrix}1&2\\3&4\end{bmatrix}$

$f = 10 e^{x+y}$

```python
from Dirichlet2D_NCN import solve_Dirichlet, plot_solve, plot_real
import numpy as np
import matplotlib.pyplot as plt
import os

c = [[
    lambda xy: 1,
    lambda xy: 2],[
    lambda xy: 3,
    lambda xy: 4
    ]]

f = lambda xy: 10 * np.exp(xy[0]+xy[1])

g = lambda xy: np.exp(xy[0]+xy[1])

xy_boundary = np.array([[0,0], [1,0], [1,1], [0,1]])


dh = 1/16

Point2Value, all_point, all_tri = solve_Dirichlet(xy_boundary, g, f, dh, c, num_epoch=1000, show_train=False)
plot_solve(Point2Value)
plt.savefig(os.path.join(".","output",f"D2NCN_solve_dh={dh}.png"))
u_real = lambda xy: np.exp(xy[0]+xy[1])
l2 = plot_real(Point2Value, u_real)
plt.savefig(os.path.join(".","output",f"D2NCN_real_dh={dh}.png"))
print(f"dh: {dh}, L2 loss: {l2}")

```

## 4

$$
u = \sum_{i} u_i(t) \Phi_i
$$


$$
<u', v> + <c\nabla u, \nabla v> = <f, v>
$$

$$
\sum u_i' <\Phi_i, \Phi_j> = <f, \Phi_j> - \sum u_i <c
\nabla \Phi_i, \nabla \Phi_j>
$$

$$
\sum \frac{u_i^{n+1} - u_i^n}{\Delta t} <\Phi_i, \Phi_j> = <f, \Phi_j> - \sum u_i <c
\nabla \Phi_i, \nabla \Phi_j>
$$

$$
\begin{aligned}
&
\sum \frac{u_i^{n+1} - u_i^n}{\Delta t} <\Phi_i, \Phi_j> 
\\&
= <\theta f^{n+1} + (1-\theta) f^n, \Phi_j> -
\\&
\sum (\theta u_i^{n+1} + (1-\theta) u_i^n) <c
\nabla \Phi_i, \nabla \Phi_j>

\end{aligned}
$$

$$
\begin{aligned}
&
\sum u_i^{n+1} (<\Phi_i, \Phi_j> +
\Delta t \theta <c \nabla \Phi_i, \nabla \Phi_j>
)
\\&
= \Delta t
<\theta f^{n+1} + (1-\theta) f^n, \Phi_j> +
\\&
\sum u_i^n (
<\Phi_i, \Phi_j> -
\Delta t \theta <c \nabla \Phi_i, \nabla \Phi_j>
)
\end{aligned}
$$

```python
from Dirichlet2D_xyt import *
import numpy as np
import matplotlib.pyplot as plt
import os

u_real = lambda xy, t: np.exp(xy[0]+xy[1]+t)

c = lambda xy, t: 2
f = lambda xy, t: - np.exp(xy[0]+xy[1]+t)

u0 = lambda xy: np.exp(xy[0]+xy[1])

g = lambda xy, t: np.exp(xy[0]+xy[1]+t)

dh = 1/8

theta = 1/2
dt = dh

xy_boundary = np.array([[0,0], [1,0], [1,1], [0,1]])
t1 = 1

Point2AllValue = ode_2D1T(c, f, g, u0, dt, t1, theta,
             dh, xy_boundary)

eval_solve(Point2AllValue, u_real)
```

# Group Work

$$
\begin{aligned}
&
u_1 = \sin (\pi x) \sin (\pi y)
\\&
u_2 = x(x-1)y(y-1)
\end{aligned}
$$


于是

$$
\begin{aligned}
&
\partial_{11} u_1 = -\pi^2 \sin(\pi x)\sin(\pi y)
\\&
\partial_{12} u_1 = \pi^2 \cos(\pi x)\cos(\pi y)
\\&
\partial_{22} u_1 = -\pi^2 \sin(\pi x)\sin(\pi y)
\end{aligned}
$$

$$
\begin{aligned}
&
\partial_{11}u_2 = 2(y^2-y)
\\&
\partial_{12}u_2 = (2x-1)(2y-1)
\\&
\partial_{22}u_2 = 2(x^2-x)
\end{aligned}
$$

$$
\begin{aligned}
&
f_1 = (\lambda+2\mu) \partial_{11}u_1 +
(\lambda+\mu)\partial_{12}u_2 +
\mu \partial_{22}u_1
\\&
f_2 = (\lambda+2\mu) \partial_{22} u_2 +
(\lambda+\mu) \partial_{12} u_1 +
\mu \partial_{11}u_2
\end{aligned}
$$

```python
from Dirichlet22 import *


g1 = lambda xy: 0
g2 = lambda xy: 0
lam = 1
mu = 2

pi = np.pi
sin = np.sin
cos = np.cos
f1 = lambda xy: (lam+3*mu) * (-pi**2 * sin(pi*xy[0]) * sin(pi*xy[1])) + (lam+mu) * (2*xy[0]-1) * (2*xy[1]-1)
f2 = lambda xy: (lam+2*mu) * 2*(xy[0]**2-xy[0]) + mu * 2*(xy[1]**2-xy[1]) + (lam+mu) * pi**2 * cos(pi*xy[0]) * cos(pi*xy[1])

xy_boundary = np.array([[0,0], [1,0], [1,1], [0,1]])
dh = 1/8

Point2Value, all_point, all_tri = solve_Dirichlet(xy_boundary, g1, g2, f1, f2, lam, mu, dh, num_epoch=7000)

plot_solve(Point2Value)

plt.savefig(os.path.join(".","output","D22_solve.png"))

u1_real = lambda xy: sin(pi*xy[0]) * sin(pi*xy[1])
u2_real = lambda xy: xy[0]*(xy[0]-1)*xy[1]*(xy[1]-1)
l2 = plot_real(Point2Value, u1_real, u2_real)
plt.savefig(os.path.join(".","output","D22_real.png"))

print(f"L2 loss: {l2}")
```
